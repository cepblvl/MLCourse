Подход 1.

1. Считываем признаки и удаляем лишние столбцы

```python
import pandas
features = pandas.read_csv('features.csv', index_col='match_id')
y=features['radiant_win']
features=features.drop(['radiant_win','duration','tower_status_radiant','tower_status_dire','barracks_status_radiant','barracks_status_dire'],axis=1)
features.head()
```

2. Проверяем пропуски


```python
a=features.count()-97230
#признаки с пропусками
print(a.index[a<0].values)
```

first_blood_time - не была совершена первая кровь за первые 5 минут

radiant_bottle_time - не была преобретена "bottle" за первые 5 минут

3. Заменяем пропуски на 0

```python
features=features.fillna(0)
features.count()
```

4. Целевую переменную содержит столбец 'radiant_win'
5. Обучение классификатора

```python
import time
import datetime
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import *
import numpy as np
kf=KFold(n_splits=5,shuffle=True,random_state=1)
kacc = []

for k in [10,20,30,40,50,60,70]:
    print(k)
    if k==30:
        start_time = datetime.datetime.now()
    clf = GradientBoostingClassifier(n_estimators=k,verbose=True,random_state=1,learning_rate=0.1,max_depth=2)
    clf.fit(np.array(features.values), np.array(y.values))
    acc = cross_val_score(estimator=clf, X=np.array(features.values), y=np.array(y.values), cv=kf, scoring='roc_auc')
    if k==30:
        time_boosting=datetime.datetime.now() - start_time
    print('end')
```

Подход 2

1. Обучаем логистическую регрессию

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
scaler = StandardScaler()
X_train=scaler.fit_transform(features)
lr = LogisticRegression(random_state=1)
grid = {'C': np.power(10.0, np.arange(-5, 6))}
kf=KFold(n_splits=5,shuffle=True,random_state=1)
gs = GridSearchCV(lr, grid, scoring='roc_auc', cv=kf)
gs.fit(X_train, y)
```

```python
bestC = gs.best_params_['C']
print(bestC)
```

```python
start_time = datetime.datetime.now()
lrC = LogisticRegression(random_state=1,C=0.01)
lrC.fit(X_train, np.array(y.values))
lracc = cross_val_score(estimator=lrC, X=X_train, y=np.array(y.values), cv=kf, scoring='roc_auc')
time_lr=datetime.datetime.now() - start_time
lraccmean=lracc.mean()
```

```python
lraccmean
```

```python
time_lr
```

2. Убираем категориальные признаки

```python
data=features.drop(['lobby_type','r1_hero','r2_hero','r3_hero','r4_hero','r5_hero','d1_hero','d2_hero','d3_hero','d4_hero','d5_hero'],axis=1)
scaler = StandardScaler()
X_train=scaler.fit_transform(data)
lr = LogisticRegression(random_state=1)
grid = {'C': np.power(10.0, np.arange(-5, 6))}
kf=KFold(n_splits=5,shuffle=True,random_state=1)
gs = GridSearchCV(lr, grid, scoring='roc_auc', cv=kf)
gs.fit(X_train, y)
bestC = gs.best_params_['C']
start_time = datetime.datetime.now()
lrC = LogisticRegression(random_state=1,C=bestC)
lrC.fit(X_train, np.array(y.values))
lracc = cross_val_score(estimator=lrC, X=X_train, y=np.array(y.values), cv=kf, scoring='roc_auc')
time_lr=datetime.datetime.now() - start_time
lraccmean=lracc.mean()
```

3. Количество идентификаторов героев

```python
heroes=features[['r1_hero','r2_hero','r3_hero','r4_hero','r5_hero','d1_hero','d2_hero','d3_hero','d4_hero','d5_hero']]
N=np.max(heroes.values)
```

4. Добавляем мешок слов по героям

```python
X_pick = np.zeros((features.shape[0], N))

for i, match_id in enumerate(features.index):
    for p in range(5):
        X_pick[i, features.ix[match_id, 'r%d_hero' % (p+1)]-1] = 1
        X_pick[i, features.ix[match_id, 'd%d_hero' % (p+1)]-1] = -1
X_train_bag=np.hstack((X_train,X_pick))
```

5. Обучаем на новых данных.

```python
lr_bag = LogisticRegression(random_state=1)
grid = {'C': np.power(10.0, np.arange(-5, 6))}
kf=KFold(n_splits=5,shuffle=True,random_state=1)
gs = GridSearchCV(lr_bag, grid, scoring='roc_auc', cv=kf)
gs.fit(X_train_bag, y)
bestC = gs.best_params_['C']
start_time = datetime.datetime.now()
lr_bagC = LogisticRegression(random_state=1,C=bestC)
lr_bagC.fit(X_train_bag, np.array(y.values))
lracc_bag = cross_val_score(estimator=lr_bagC, X=X_train_bag, y=np.array(y.values), cv=kf, scoring='roc_auc')
time_lr_bag=datetime.datetime.now() - start_time
lraccmean_bag=lracc_bag.mean()
```

```python
lraccmean_bag
```
